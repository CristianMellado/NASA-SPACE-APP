```python
# YOLOv5
!pip install -q torch torchvision torchaudio
!git clone https://github.com/ultralytics/yolov5.git
%cd yolov5
!pip install -qr requirements.txt

import torch
from IPython.display import Image, display

#trained model
model_path = '/content/drive/MyDrive/NASAA/craters/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

print("Clases detectadas:", model.names)
```

```python
# predicciones
img_path = '/content/drive/MyDrive/NASAA/craters/pruebas/crater1luna.jpeg'

results = model(img_path)
results.save()

import glob, os
from IPython.display import Image, display

last_exp = sorted(glob.glob('runs/detect/exp*'))[-1]

result_images = glob.glob(os.path.join(last_exp, '*.*'))

if len(result_images) == 0:
    print("‚ö†Ô∏è No se encontr√≥ ninguna imagen resultante en:", last_exp)
else:
    print(f"üìÅ Mostrando imagen detectada desde: {last_exp}")
    display(Image(filename=result_images[0]))
```
